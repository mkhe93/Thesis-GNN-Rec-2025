## Evaluate Hyper-Parameter Search

- use `HyperSearchResults.csv` to plot hyper-parameters against target metric ($NDCG$)

```python
import pandas as pd
from src.config import HYPER_DIRECTORY
from utils import plot_parameter_influence, plot_parameter_influence_knn
```

```python
file_path = HYPER_DIRECTORY.joinpath("results/HyperSearchResults.csv")
hyper_search_results = pd.read_csv(file_path, sep="\t")
```

#### choose `type` between `scatter` and `line` in the following

```python
parameters = [['alpha', '$\\alpha$'], ['beta', '$\\beta$'], ['k', '$k$'], ['q', '$q$']]

selected_algorithm = ['asymuserknn', 'asymitemknn']
labels = {'asymitemknn' : 'AsymItemKNN', 'asymuserknn': 'AsymUserKNN'}

plot_parameter_influence_knn(
    hyper_search_results,
    parameters=parameters,
    metric=['test_ndcg@10', '$NDCG@10$'],
    algorithms=selected_algorithm,
    custom_labels=labels,
    type="scatter",
    save_fig=True
)
```
